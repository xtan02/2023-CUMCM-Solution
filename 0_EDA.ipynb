```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 设置中文字体为宋体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 设置英文字体为新罗马
plt.rcParams['font.serif'] = ['Times New Roman']
# 字体大小
plt.rcParams['font.size'] = 12  
# 正常显示负号
plt.rcParams['axes.unicode_minus'] = False 
```

## ***探索性可视化分析***

```python
fig, ax = plt.subplots(1,2,figsize=(20, 3))

sns.countplot(x='分类编码', palette='Set2', data=df1, ax=ax[0])
sns.countplot(x='分类名称', palette='Set2', data=df1, ax=ax[1])
```

```python
fig, ax = plt.subplots(1,2,figsize=(20, 3))

sns.countplot(x='销售类型', palette='Set2', data=df2, ax=ax[0])
sns.countplot(x='是否打折销售', palette='Set2', data=df2, ax=ax[1])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(25, 5))
sns.countplot(x='分类名称', palette='Set2', data=df1, ax=ax[0])
sns.countplot(x='销售类型', palette='Set2', data=df2, ax=ax[1])
sns.countplot(x='是否打折销售', palette='Set2', data=df2, ax=ax[2])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 3))
sns.histplot(data=df2, x="销量(千克)", binwidth=4, kde=True, ax=ax[0])
sns.histplot(data=df2, x="销售单价(元/千克)", binwidth=4, kde=True, ax=ax[1])
```
