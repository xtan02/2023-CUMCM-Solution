```python
import pandas as pd
```

```python
df1 = pd.read_excel("data\\附件1.xlsx").drop(["单品名称", "分类编码"], axis=1)
df1.head()
```

```python
df2 = pd.read_csv("data\\附件2.csv")
df2.head()
```

```python
df3 = pd.read_excel("data\\附件3.xlsx")
df3.columns = ["销售日期", "单品编码", "批发价格(元/千克)"]
df3.head()
```

```python
df4 = pd.read_excel("data\\附件4.xlsx"#, sheet_name="Sheet1"
                    )
df4.columns = ["分类编码", "分类名称", "平均损耗值"]
df4.head()
```

```python
df1['单品编码'] = df1['单品编码'].astype(str)
df2['销售日期'] = df2['销售日期'].astype(str)
df2['单品编码'] = df2['单品编码'].astype(str)
df3['销售日期'] = df3['销售日期'].astype(str)
df3['单品编码'] = df3['单品编码'].astype(str)
```

```python
df = df2.merge(df3, on=['销售日期', '单品编码'], how='left').drop(["销售类型", "是否打折销售"], axis=1)
df.head(10)
```

```python
df_final1 = pd.merge(df, df1, on="单品编码", how="left")
df_final1.head()
```

```python
df_final = pd.merge(df_final1, df4, on="分类名称", how="left")
df_final.head()
```

```python
def devide(df, name="分类名称"):
    # 获取分类特征的所有唯一值
    categories = df[name].unique()

    # 创建一个空的字典，用于存储划分后的DataFrame
    dfs_by_category = {}

    # 按分类特征划分DataFrame
    for category in categories:
        dfs_by_category[category] = df[df[name] == category]

    # 将每个划分后的DataFrame写入CSV文件
    for category, df_category in dfs_by_category.items():
        if name == "分类名称":
            csv_filename = f'data_2\{category}.csv'  
        else:
            csv_filename = f'data_2\{category}.csv'  
        df_category.to_csv(csv_filename, encoding='gbk', index=False)  
```

```python
devide(df_final)
```
