```python
import pandas as pd
```

```python
df = pd.read_csv("data\\附件2.csv")
df['销售日期'] = pd.to_datetime(df['销售日期'])
# 设置时间戳列为索引（如果尚未设置）
df.set_index('销售日期', inplace=True)
df['日期'] = df.index

# 选择2023年6月24日到30日的数据
start_date = '2023-06-24'
end_date = '2023-06-30'
selected_data = df[start_date:end_date]
selected_data.head()
```

```python
vegetable_names = selected_data["单品编码"].tolist()
# 使用集合（set）来去除重复项
vegetable_names = set(vegetable_names)
# 将集合转换回列表
vegetable_names = list(vegetable_names)
len(vegetable_names)
```

```python
df3 = pd.read_excel("data\\附件3.xlsx")
df3.columns = ["日期", "单品编码", "批发价格(元/千克)"]
```

```python
df_1 = selected_data.merge(df3, on=['日期', '单品编码'], how='left')
df_1.head(10)
```

```python
df4 = pd.read_excel("data\\附件4.xlsx", sheet_name="Sheet1")
df4.head()
```

```python
df_2 = df_1.merge(df4, on=['单品编码'], how='left').drop(["扫码销售时间", "销售类型", "是否打折销售", "单品编码"], axis=1)
df_2.head()
```

```python
def devide(df):
    # 获取分类特征的所有唯一值
    categories = df["单品名称"].unique()

    # 创建一个空的字典，用于存储划分后的DataFrame
    dfs_by_category = {}

    # 按分类特征划分DataFrame
    for category in categories:
        dfs_by_category[category] = df[df["单品名称"] == category]

    # 将每个划分后的DataFrame写入CSV文件
    for category, df_category in dfs_by_category.items():
        csv_filename = f'data_3\{category}.csv'  
        df_category.to_csv(csv_filename, encoding='gbk', index=False)  

devide(df_2)
```
