```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel("data\\附件1.xlsx").drop(["分类编码"], axis=1)
df1.head()
```

```python
df2 = pd.read_csv("data\\附件2.csv")
df2.head()
```

```python
df3 = pd.merge(df1, df2, on='单品编码', how='right').drop("单品编码", axis=1)
df3.head()
```

```python
def devide(df, name="分类名称"):
    # 获取分类特征的所有唯一值
    categories = df[name].unique()

    # 创建一个空的字典，用于存储划分后的DataFrame
    dfs_by_category = {}

    # 按分类特征划分DataFrame
    for category in categories:
        dfs_by_category[category] = df[df[name] == category]

    # 将每个划分后的DataFrame写入CSV文件
    for category, df_category in dfs_by_category.items():
        if name == "分类名称":
            csv_filename = f'data_1\按类别划分附件2\{category}.csv'  
        else:
            csv_filename = f'data_1\按单品划分附件2\{category}.csv'  
        df_category.to_csv(csv_filename, encoding='gbk', index=False)  
```

```python
devide(df3)
devide(df3, "单品名称")
```
